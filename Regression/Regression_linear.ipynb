```python
q_df=q_df[['Adj. Open','Adj. High','Adj. Low','Adj. Close','Adj. Volume']]
```

```python
q_df['HGH_PCT']=(q_df['Adj. High']-q_df['Adj. Close'])*100/q_df['Adj. Close']
```

```python
q_df['PCT_CHG']=(q_df['Adj. Close']-q_df['Adj. Open'])*100/q_df['Adj. Open']
```

```python
q_df=q_df[['Adj. Close','HGH_PCT','PCT_CHG','Adj. Volume']]
```

```python
q_df
```

*Identified key features, explored the relationship between different features and prepared the data frame likewise*

```python
forecast_col='Adj. Close'
```

```python
q_df.fillna(-99999,inplace=True)
forecast_out=int(math.ceil(0.01*len(q_df)))
```

```python
df=q_df
df['label']=df[forecast_col].shift(-forecast_out)
#to relate the features to the prediciton by 36 days(based on the data that we have rn) before.
```

```python
df.dropna(inplace=True)
#to drop the rows by default that contain atleast one na value
```

```python
df.head(36)
```

```python
forecast_out
```

<u>Time to train and test the model</u>

```python
import numpy as np
```

```python
from sklearn import preprocessing,model_selection,svm
from sklearn.linear_model import LinearRegression
```

```python
X = np.array(df.drop(['label'],axis=1))
#here the frist parameter is the cokumn that I am dropping and the second one is axis which when "1" indicates that a column is selected not a row
y = np.array(df['label'])
#we set the output labels
X = preprocessing.scale(X)
X= X[:-forecast_out]
X_lately=X[-forecast_out:]
'''
here we are normalinsing the values of X's features against all the other values for the ease of computation against the training data as well but for
high speed applications such as fast trading and stuff we avoid this process since this can take processing time to increase while computation time dec
'''
df.dropna(inplace=True)
y = np.array(df['label'])
y_lately=y[-forecast_out:]
y=y[:-forecast_out]
```

```python
print(len(X),len(y))
```

```python
X_train,X_test,y_train,y_test=model_selection.train_test_split(X,y,test_size=0.2)
```

```python
print(f"{X_train}\ntest\n{X_test}")
```

```python
clf = LinearRegression()
clf.fit(X_train,y_train)
#fit is to fit a graph through the training data
accuracy = clf.score(X_test,y_test)
#score is to check the accuracy by testing using the testing data
with open('linear_regression.pickle','wb') as f:
    pickle.dump(clf,f)
pickle_in= open('linear_regression.pickle','rb')
clf=pickle.load(pickle_in)
if not pickle_in.closed:
    pickle_in.close()
accuracy
```

*Let's compare that with a Supprt Vector Machine*

```python
accuracy = clf.score(X_test,y_test)
yaclf=svm.SVR()
yaclf.fit(X_train,y_train)
svmAccuracy = yaclf.score(X_test,y_test)
accuracy
```

```python
forecast_set = clf.predict(X_lately)
print(f"forecast is:\n{forecast_set} \n\nand the actual values were:\n{y_lately}")
```
